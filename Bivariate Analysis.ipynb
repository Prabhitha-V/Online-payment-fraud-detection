```python
# Importing Necessary Libraries

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import scipy.stats as st
from sklearn.utils import resample
from xgboost import XGBClassifier
from itertools import combinations as comb

%matplotlib inline
print("Imported Sucessfully.")
```

```python
# Loading The Dataset

opfd=pd.read_csv("C:/Users/deeks/Downloads/onlinefraud.csv", encoding="latin1")
print("Online Fraud Detection Data Loaded Successfully.")
```

```python
# Ignore Warnings

import warnings
warnings.filterwarnings('ignore')

# loss Due To Fraud Transaction

print("The amount of loss due to fraud transaction:")
opfd[opfd.isFraud==1]['amount'].sum()

# Counts

opfd[['type', 'isFraud']].value_counts()

# Plotting Distplot

plt.figure(figsize=(15, 6))
sns.distplot(opfd['step'], bins=50, color ="pink")
```

```python
# Bivariate Analysis For Numerical

opfd2 = resample(opfd,
              random_state=42,
              n_samples=10000,
              stratify=opfd['isFraud']).reset_index(drop=True)

opfd2_num = opfd2.select_dtypes('number').columns
norm_data = []
y=1
high_cor = []
low_cor = []

for col1, col2 in comb(opfd2_num,2):
    if (col1 != 'Credit_Score') or (col2 != 'Credit_Score'):
        plt.figure(figsize=(5,5))
        print(f'{y}. {col1} - {col2}')
        opfd2_na = opfd2.dropna()
        sns.regplot(data=opfd2_na, x=col1, y=col2, line_kws={'color':'red'})
        plt.tight_layout()
        plt.show()
        if (col1 in norm_data) & (col2 in norm_data):
            pear, pval1 = st.pearsonr(opfd2_na[col1],opfd2_na[col2])
            print(f'data have pearson correlation with spesification:')
            print(f'corr: {pear}')
            print(f'pvalue: {pval1}')
            if (pear > 0.8) | (pear < -0.8):
                high_cor.append([col1,col2])
                print(f'High corr between {col1} and {col2}')
            else:
                low_cor.append([col1,col2])
                print(f'Low corr between {col1} and {col2}')
        else:
            spear, pval2 = st.spearmanr(opfd2_na[col1],opfd2_na[col2])
            print(f'data have spearman correlation with spesification:')
            print(f'corr: {spear}')
            print(f'pvalue: {pval2}')
            if (spear > 0.9) | (spear < -0.9):
                high_cor.append([col1,col2])
                print(f'High corr between {col1} and {col2}')
            else:
                low_cor.append([col1,col2])
                print(f'Low corr between {col1} and {col2}')
        print('')
        y+=1

print('')
```
