```python
# Importing Necessary Libraries

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.utils import resample

%matplotlib inline
print("Imported Sucessfully.")
```

```python
# Loading The Dataset

opfd=pd.read_csv("C:/Users/deeks/Downloads/onlinefraud.csv", encoding="latin1")
print("Online Fraud Detection Data Loaded Successfully.")
```

```python
# Ignore Warnings

import warnings
warnings.filterwarnings('ignore')

#Fraud Index And fFraud Values

fraud = opfd['isFraud'].value_counts()
fraud_idx = fraud.index
fraud_val = fraud.values
```

```python
# Plotting Pie Chart For Fraud_Value And Fraud_Index

plt.pie(fraud_val,
        labels=fraud_idx,
        autopct='%1.1f%%',
        startangle=90)
plt.show()
```

```python
# Dropping Nan Values

opfd.dropna(inplace=True)
```

```python
# Resampling opfd And Relabelling As opfd2

opfd2 = resample(opfd,
              random_state=42,
              n_samples=10000,
              stratify=opfd['isFraud']).reset_index(drop=True)
```

```python
opfd2.head()

# Finding Index And Values For Resampled Dataset

fraud = opfd2['isFraud'].value_counts()
fraud_idx = fraud.index
fraud_val = fraud.values

# Plotting A Pie Chart For Fraud Index And Values

plt.pie(fraud_val,
        labels=fraud_idx,
        autopct='%1.1f%%',
        startangle=90)
plt.show()
```
